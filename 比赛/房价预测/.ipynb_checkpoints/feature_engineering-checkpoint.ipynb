#特征选择

1、特征是否发散，也就是方差的大小。如果方差太小的话，他包含的信息就不够多。

2、看相关性大小

特征选择方法有三种：

1、Filter过滤法：按照特征的发散性（方差大小）、相关性来排序

2、Wrapper包装法：根据后期的学习器，也就是目标函数的评分，每次选取若干的特征，或者排除若干的特征

3、Embedded嵌入法：上面两者的结合，先使用某些机器学习算法进行训练，得到各个特征的权值系数，然后再排序进行选择

或者用sklearn中的feature_selection库来进行特征选择

##Filter 方差选择法

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
var = VarianceThreshold(2) #方差低于3的特征直接删除
```

```python
var_filter = var.fit_transform(data.data)
```

##单变量特征选择

考虑自变量等于i且因变量等于j的样本频数的观察值与期望值的差距

sum((实际值-理论值)/理论值）

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import chi2
```

```python
SelectKBest(chi2, k=2).fit_transform(data.data, data.target) # 选择变量和标签之间的相关性
```

##皮尔逊相关系数

他只能衡量线性相关，不能衡量非线性相关

##Wrapper包装法

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
RFE(estimator=LogisticRegression(), n_features_to_select=2).fit_transform(data.data, data.target)
# 基模型是逻辑回归，选择特征2个
```

##Embedded 嵌入法
从模型的角度来进行选择，带有惩罚

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
SelectFromModel(LogisticRegression(penalty="l2",C=0.1)).fit_transform(data.data, data.target)
#l2
```

##基于树模型的特征选择

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
SelectFromModel(GradientBoostingClassifier()).fit_transform(data.data, data.target)
```
